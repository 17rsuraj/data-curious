```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dummy_Sales_Data_v1.csv")
df.head()
```

```python
#Filtering using single condition
df.query("Quantity == 95")
```

```python
#Filtering using two condition in AND logic
df.query("Quantity == 95 and UnitPrice(USD) == 182")
```

```python
#Filtering using two condition in AND logic
df.query("Quantity == 95 and `UnitPrice(USD)` == 182")
```

```python
df.rename(columns={'UnitPrice(USD)':'UnitPrice', 
                   'Shipping_Cost(USD)':'Shipping_Cost',
                   'Delivery_Time(Days)':'Delivery_Time'},
          inplace=True)
df.head()
```

```python
#Filtering using two condition in AND logic
df.query("Quantity == 95 and UnitPrice == 182")
```

```python
#Filtering using two condition in OR logic
df.query("Quantity == 95 or UnitPrice == 182")
```

```python
df.query("Quantity == 95 | UnitPrice == 182")
```

```python
#Filtering using one condition in NOT logic
df.query("not (Quantity == 95)")
```

```python
df.query("~(Quantity == 95)")
```

```python
df.query("Quantity != 95")
```

```python
# Filtering on Text column
df.query("Status == 'Not Shipped'")
```

```python
# Filtering with math operation
df.query("Shipping_Cost*2 < 50")
```

```python
# Filtering with math operation
df.query("Quantity**2 + Shipping_Cost**2 < 500")
```

```python
df[df["Quantity"]**2 + df["Shipping_Cost"]**2<500]
```

```python
#Filtering using one condition and Maths calculation in it
df.query("UnitPrice > Shipping_Cost*5")
```

```python
#Filtering using one condition and Maths calculation in it with built-in function
df.query("sqrt(UnitPrice) > 15")
```

```python
#Filtering using one condition and Maths calculation in it with built-in function --> combination of maths and function
df.query("sqrt(UnitPrice) < Shipping_Cost/2")
```

```python
#Filtering using Text values
df.query("Quantity == 95 and Product_Category == 'Healthcare'")
```
