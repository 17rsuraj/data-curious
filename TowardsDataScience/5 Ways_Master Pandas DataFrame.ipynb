```python
import faker
import pandas as pd
import numpy as np
```

```python
name = ['Ron', 'Jacob', 'Rafael', 'Karl-Hans', 'Freddy', 'Ron', 'Jacob']
age =[30,24,83,34,20,55,29]
height = [153, 153, 161, 169, 169, 172, 178]
weight = [69, 89, 50, 69, 86, 85, 63]
shirt_size = ['S', 'M', 'M', 'L', 'S', 'L', 'L']

#Make DataFrame
df = pd.DataFrame({"name": name,
                  "age": age,
                  "height": height,
                  "weight": weight,
                  "shirt_size": shirt_size})
df.head(10)
```

```python
df1 = df.sort_values("age")
df1.head(10)
```

```python
df2 = df.sort_values("age", ascending=False)
df2.head(10)
```

```python
df3 = df.sort_values(["weight", "height"])
df3.head(10)
```

```python
df3 = df.sort_values(["height", "weight"])
df3.head(10)
```

```python
df4 = df.sort_values(["weight", "height"], ascending = [True, False])
df4.head(10)
```

```python
df6 = df.sort_values("age", ascending=False)
df6.head(10)
```

```python
df5 = df.sort_values("age", na_position = 'first')
df5.head(10)
```

```python
df_idx = pd.DataFrame({"shirt_size": ['S','M','L']})  
idx = df_idx.reset_index().set_index("shirt_size")    
idx
```

```python
df["shirt_size_code"] = df["shirt_size"].map(idx["index"])
df.head(10)
```

```python
df.sort_values("shirt_size_code")
df.head(10)
    
```
